```python
import pandas as pd
import numpy as np 
import itertools
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
average_expression_per_tissue = pd.read_csv('/home/passala/Gene_duplication_project/Code/Whole_genome_comparison_code/Temp_tables/avg_expression_per_gene_all_tissue_types.csv', index_col = 0)
average_expression_per_tissue
```

```python
primary_dataset_cols = []
secondary_dataset_cols = []
tertiary_dataset_cols = []
quatenary_dataset_cols = []
for col in average_expression_per_tissue.columns.to_list():
    if 'Dataset 1' in col:
        primary_dataset_cols.append(col)
    if 'Dataset 2' in col:
        secondary_dataset_cols.append(col)
    if 'Dataset 3' in col:
        tertiary_dataset_cols.append(col)
    if 'Dataset 4' in col:
        quatenary_dataset_cols.append(col)
```

```python
average_expression_per_tissue[primary_dataset_cols]
```

```python
dataset_list = [primary_dataset_cols,secondary_dataset_cols,tertiary_dataset_cols,quatenary_dataset_cols]
list_of_set_names = ['Dataset 1','Dataset 2','Dataset 3','Dataset 4']
```

```python
correlation_results_dict = {}
for dataset_cols,dataset_names  in zip(dataset_list,list_of_set_names):
    sub_setted_genes = average_expression_per_tissue[dataset_cols]
    current_corr_frame = sub_setted_genes.T.corr(method = 'spearman')
    current_corr_frame = current_corr_frame.replace(1,0)
    current_corr_frame = current_corr_frame.dropna(axis = 0, how = 'all')
    current_corr_frame = current_corr_frame.dropna(axis = 1, how = 'all')
    correlation_results_dict[dataset_names] = current_corr_frame
    
```

```python
all_shared_expressed_genes = set(correlation_results_dict['Dataset 1'].index.to_list()).intersection(set(correlation_results_dict['Dataset 2'].index.to_list()))
all_shared_expressed_genes = all_shared_expressed_genes.intersection(correlation_results_dict['Dataset 3'].index.to_list())
all_shared_expressed_genes = all_shared_expressed_genes.intersection(correlation_results_dict['Dataset 4'].index.to_list())

for current_corr in list_of_set_names:
    get_df = correlation_results_dict[current_corr]
    updated_df = get_df[get_df.columns.intersection(all_shared_expressed_genes)]
    updated_df = updated_df[updated_df.index.isin(all_shared_expressed_genes)]
    correlation_results_dict[current_corr] = updated_df
    
```

```python
ranked_correlation_results ={}
for current_corr in list_of_set_names:
    get_df = correlation_results_dict[current_corr]
    ranked_df = get_df.rank()
    ranked_correlation_results[current_corr] = ranked_df
```

```python
for combo in all_possible_combos:
    current_base_correlation = correlation_results_dict[combo[0]]
    target_ranked_dataset = ranked_correlation_results[combo[1]]
    col_name = f"{combo[0]} To {combo[1]}" 

    top_10_arab_genes =  [current_base_correlation[c].nlargest(10).index.values for c in current_base_correlation] # using pair list above, cut down top 10 list to relevant genes, probably by adding list as a column in panda and then filtering panda to index of pair list
    top_10_arab_genes_dataframe = pd.DataFrame(
        data=top_10_arab_genes,
        index=current_base_correlation.index,
        columns=[
            "One",
            "Two",
            "Three",
            "Four",
            "Five",
            "Six",
            "Seven",
            "Eight",
            "Nine",
            "Ten",
        ],
    )
    top_10_arab_genes_dataframe

    for current_base_gene in top_10_arab_genes_dataframe.index:
        list_of_fingerprint_genes = top_10_arab_genes_dataframe.loc[current_base_gene].to_list()
        ranks_in_target_df = target_ranked_dataset.loc[target_ranked_dataset.index.isin(list_of_fingerprint_genes), current_base_gene]
        mean_rank = ranks_in_target_df.mean()
        results_of_pairwise_comparison.at[current_base_gene,col_name] = mean_rank
        
    
    
    
    
```
